```julia
using ForwardDiff
using PyPlot
```

```julia
f(x)=exp(x)+cos(0.3)*sin(9x)

fDf(x, h) = (f(x+h)-f(x))/h
cDf(x, h) = (f(x+h)-f(x-h))/2h

Df(x)=ForwardDiff.derivative(f, x)
```

```julia
x=1.2
h=[10.0^(-k) for k in 1:25]

abserr_cDf=  abs.(Df(x) .- cDf.(x,h))
abserr_fDf=  abs.(Df(x) .- fDf.(x,h))


loglog(h, abserr_cDf, h, abserr_fDf)
legend(["central differencing", "forward differencing"])
```
